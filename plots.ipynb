```python
# create a list for all values
humans = []
neighbours = []

# write values to list
j=1
with open('no_alignment_no_cohesion.txt', 'r') as f:
    for line in f.readlines()[1:]:
        l = line.strip().split(',')
        density = []
        speed = []
        humans.append(int(l[0]))
        neighbours.append(int(l[1]))
        length = int(l[2])
        for i in range(3,3+length,1):
            density.append(float(l[i]))
        for i in range(3+length,3+2*length,1):    
            speed.append(float(l[i]))
        fig = plt.plot(figsize=(8,6))
        #legend of dist
        #density = np.mean(density)
        #speed = np.mean(speed)
        plt.scatter(density, speed, label = j)
        j+=1
plt.xlabel('density (humans/m2)', size=10)
plt.ylabel('speed (m/timestep)', size=10)
plt.title('FD with speed and density using only separation', size=12)

plt.ylim([0, 2.5])
plt.grid()
plt.legend(mode = 'expand', ncol = 5, title = 'Experiment')
#plt.savefig("FD_noAl_noCo.jpg")
plt.show()
```
